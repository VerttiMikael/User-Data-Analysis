## 3rd chart

### Active and Inactive users

#### Pie chart with stacked bar chart


```python
# inspecting the data that we have for the pie bar chart

df1['REGISTRATION_COUNTRY'].value_counts()
```

```python
# -- ACTIVE AND INACTIVE USERS + COUNTRY OF ORIGIN --


# making a figure and assigning axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

# pie chart parameters
size1 = df1.shape[0]
size2 = df2.shape[0]
sizes = [size1, size2]
labels = f'Active ({size1} users)', f'Inactive ({size2} users)'
explode = [0.1, 0,]
ax1.set_title('active and inactive users')

# rotating the pie
angle = 90 * sizes[1]
wedges, *_ = ax1.pie(sizes, autopct='%1.1f%%', startangle=angle,
                     labels=labels, explode=explode, 
                     colors=['steelblue', 'darkorange'], shadow=True)

# bar chart parameters
countries = df1['REGISTRATION_COUNTRY'].value_counts()
bottom = 1
width = .2

# top 3 countries and combine the rest into "Other"
top_countries = countries.head(3)
other_count = countries.iloc[3:].sum()
top_countries.loc['Other'] = other_count

# creating the bar chart
for j, (height, label) in enumerate(reversed([*zip(top_countries, top_countries.index)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='steelblue', label=label,
                 alpha=0.2 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height}"], label_type='center')

ax2.set_title('Top registration countries')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

#--DRAWING THE LINES--

# using ConnectionPatch to draw lines between the two plots
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(top_countries.values)



# top connecting line
bar_height = sum(top_countries.values)
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(1)
ax2.add_artist(con)

# bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, -bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(1)

plt.show()
```

## 4th chart:

### The most preferred cuisines

#### Horizontal bar chart

```python
# Inspecting how the preferred restaurants are stored

df1['PREFERRED_RESTAURANT_TYPES'].value_counts()
```

```python
# -- PREFERRED RESTAURANTS TYPES-- 


# -- MODIFYING THE PREFERRED RESTAURANTS DATA SO THAT WE CAN USE IT --

# Creating a copy of the dataframe so it doesnt affect the original data
df_copy = df1.copy()

# Removing '\n' characters and whitespaces, also dropping rows with 'NaN' values
df_copy.loc[:, 'PREFERRED_RESTAURANT_TYPES'] = \
df_copy['PREFERRED_RESTAURANT_TYPES'].str.replace(r'"|\n\s*', '', regex=True)

restaurants = df_copy.dropna(subset=['PREFERRED_RESTAURANT_TYPES'])

# Splitting the values into a list and explode because of the rows that have more than one value in them
restaurants_exploded = restaurants['PREFERRED_RESTAURANT_TYPES'].apply(lambda x: 
    [c.strip() for c in x.strip('[]').split(',') if pd.notna(x)]
    ).explode()


# -- MAKING THE CHART -- 

ax = restaurants_exploded.value_counts().plot(kind='barh', 
                                        color='orangered',
                                        alpha=0.8,
                                        title= 'The most preferred cuisines')

ax.grid(False)
ax.invert_yaxis()
ax.tick_params(colors='black', which='both') 

```

## 5th chart:

### The most popular food types

#### bar chart

```python
# -- WHATS THE MOST POPURAL FOOD TYPE -- 

breakfast = df1['BREAKFAST_PURCHASES'].sum()
lunch = df1['LUNCH_PURCHASES'].sum()
eveningsnack = df1['EVENING_PURCHASES'].sum()
dinner = df1['DINNER_PURCHASES'].sum()
latenight = df1['LATE_NIGHT_PURCHASES'].notna().sum()
all_foods = [breakfast, lunch, eveningsnack, dinner, latenight]
foods_names = ['Breakfast', 'Lunch', 'Evening Snack', 'Dinner', 'Late Night']



plt.bar(foods_names, all_foods, color='steelblue')
plt.title('The most popular food types')
plt.ylabel('Total Purchases')
plt.show()
```

## 6th chart:

### Different purchase statistics for every preferred device

#### Bar and Pie charts

```python
# -- DOES APPLE USERS SPEND MORE THAN ANDROID OR WEB USERS? --


# Making a figure and assigning axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(wspace=0.3)

# Grouping the dataframe by device type and calculating the mean purchase value and total purchase value
grouped = df1.groupby('PREFERRED_DEVICE').agg({'AVG_PURCHASE_VALUE_EUR': 'mean', 'PURCHASE_COUNT': 'sum'})
grouped['TOTAL_PURCHASE_VALUE_EUR'] = grouped['AVG_PURCHASE_VALUE_EUR'] * grouped['PURCHASE_COUNT']

# Creating the bar chart
ax1.bar(grouped.index, grouped['AVG_PURCHASE_VALUE_EUR'], color='steelblue')
ax1.set_xlabel('Device Type')
ax1.set_ylabel('Average Purchase Value (EUR)')
ax1.set_title('Average Purchase Value')

# Creating the pie
labels = grouped.index
sizes = grouped['TOTAL_PURCHASE_VALUE_EUR']
max_index = np.argmax(sizes)
explode = [0.1 if i == max_index else 0 for i in range(len(labels))]
ax2.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90,
        shadow=True, colors=['lightgreen', 'indianred', 'steelblue'])
ax2.axis('equal')
ax2.set_title('Percentage of The Total Purchase Value')

# Creating a box that showcases the total value committed
committed_values = grouped['TOTAL_PURCHASE_VALUE_EUR'] / 1000 # convert to thousands
ax2.text(1.3, -0.8, 'Total value:', fontsize=10, ha='right')
for i, device in enumerate(grouped.index):
    ax2.text(1.3, -1.0 - 0.2*i, f'{device}: {committed_values[i]:,.0f}k', fontsize=10, ha='right')

plt.show()

```
